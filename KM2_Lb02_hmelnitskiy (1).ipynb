Копирование объектов возможно с помощью вызова соответствующих функций-конструкторов.

```python
obj = [1, 4, 10]; print(obj, id(obj))
obj_copy = list(obj); obj_copy = obj.__class__(obj) 
print(obj_copy, id(obj_copy))
```

Копирование последовательностей ( list , str , tuple ) возможно посредством операции среза по всем элементам [:]

```python
objs2 = 'String', [54, 234, 3], (0,9,7)
objs2_copy = tuple(map(lambda x: x[:], objs2))
objs2_copy
```

Копирование изменяемых объектов (списков ( list ), словарей ( dict ) и множеств ( set )) возможно с помощью метода copy

```python
objs3 = ([1, 2, 3], {'a':1, 'b':2, 'c':3}, {1,4,'s',(5,)})
objs3_copy = tuple(map(lambda x: x.copy(), objs3))
objs3_copy
```

#### Задание 2.12 Файловый объект

Приведите пример, в котором используются следующие функции и методы для работы с файловым объектом: open, close, read, readline, readlines, write, writelines.

```python
file = open('example.txt', 'w')

file.write('Hello\n')
file.writelines(['This is line 2.\n', 'This is line 3.\n'])

file = open('example.txt', 'r')

print('file.read()')
print(file.read())

file.seek(0)

print('file.readline()')
print(file.readline())

file.seek(0)

print('file.readlines()')
print(file.readlines())

file.close()
```
