## Задание 4.1. Построение треугольника Рело

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 3
r = 10
N = 100
center = np.array([0,0])
```

```python
l = r
```

```python
R = l/ (2*np.sin(np.pi/n))
```

```python
t = np.arange(0, 2 * np.pi, 2/3 * np.pi)
```

```python
vertices = center + R * np.transpose([np.cos(t), np.sin(t)])
```

**Последовательность вычислений для выражения**

`vertices = center + R * np.transpose([np.cos(t), np.sin(t)])`

1. **`t`** — массив углов (например, `np.linspace(0, 2π, N)`).
2. **`[np.cos(t), np.sin(t)]`** — список из двух массивов: косинусы и синусы углов.
3. **`np.transpose(...)`** — преобразует список `[массив_косинусов, массив_синусов]` в массив формы `(N, 2)`, где каждая строка — `(cos, sin)` для одного угла.
4. **`R * ...`** — поэлементное умножение на радиус `R`, масштабирование координат.
5. **`center + ...`** — прибавление координат центра (сдвиг).
6. **`vertices`** — итоговый массив формы `(N, 2)` с координатами вершин многоугольника/окружности.

```python
plt.plot(vertices[:,0], vertices[:,1],'b.', markersize = 30)
plt.axis('equal')
```

```python
alpha = 2*np.pi / n
beta = alpha / 2
```

```python
angle = np.linspace(-beta / 2, beta / 2, N)
```

```python
side0 = vertices[0] + r * np.transpose([np.cos(np.pi + angle), np.sin(np.pi + angle)])
side1 = vertices[1] + r * np.transpose([np.cos(np.pi + angle + alpha), np.sin(np.pi + angle + alpha)])
side2 = vertices[2] + r * np.transpose([np.cos(np.pi + angle + 2* alpha), np.sin(np.pi + angle + 2 * alpha)])
```


\begin{cases}
x(angle) = x_0 + r \cdot \cos(\pi + angle) \\
y(angle) = y_0 + r \cdot \sin(\pi + angle)
\end{cases}


```python
plt.plot(vertices[:,0], vertices[:,1],'b.', markersize = 20)
plt.plot(side0[:,0], side0[:,1],'g-', markersize = 20)
plt.plot(side1[:,0], side1[:,1],'g-', markersize = 20)
plt.plot(side2[:,0], side2[:,1],'g-', markersize = 20)
plt.axis('equal')
```

```python
list = [vertices[i] + r * np.transpose([np.cos(np.pi + angle + i*alpha), np.sin(np.pi + angle + i*alpha)]) for i in range(n)]
```

```python
sides = np.concatenate(list)
```

```python
plt.plot(vertices[:,0], vertices[:,1],'b.', markersize = 20)
plt.plot(sides[:,0], sides[:,1],'g-', markersize = 20)
plt.axis('equal')
```

## Задание 4.2. Построение правильного многоугольника Рело


```python
n = 5
r = 5
N = 100
center = np.array([1,0])
```

```python
l = (2*(r**2) * (1 - np.cos(np.pi / n)))**0.5
```

 Вывод формулы стороны правильного многоугольника  
*( $n$ — нечётное )*

**Дано:**  
- $n$ — нечётное количество вершин  
- $r$ — длина самой большой диагонали  
- $L$ — сторона правильного многоугольника  

---

 1. Связь $r$ с радиусом описанной окружности $R$

Центральный угол между вершинами самой большой диагонали:

$$
\theta = \frac{\pi}{n}
$$

По теореме косинусов:

$$
l^2 = r^2 + r^2 - 2r^2 \cos\theta
$$

---

 4. Итоговая формула

$$
\boxed{l = (2r^2 (1 - \cos\theta))^\frac{1}{2} }
$$


```python
R = l/ (2*np.sin(np.pi/n))
```

```python
t = np.arange(0, 2 * np.pi, 2 * np.pi / n)
```

```python
vertices = center + R * np.transpose([np.cos(t), np.sin(t)])
```

```python
plt.plot(vertices[:,0], vertices[:,1],'b.', markersize = 20)
plt.axis('equal')
```

```python
alpha = 2 * np.pi / n
beta = alpha / 2
angle = np.linspace(-beta / 2, beta / 2, N)
```

```python
list = [vertices[i] + r * np.transpose([np.cos(np.pi + angle + alpha * i), np.sin(np.pi + angle + alpha * i)]) for i in range(n)]
sides = np.concatenate(list)
```

```python
plt.plot(vertices[:,0], vertices[:,1],'b.', markersize = 20)
plt.plot(sides[:,0], sides[:,1],'k-', markersize = 20)
plt.axis('equal')
```
