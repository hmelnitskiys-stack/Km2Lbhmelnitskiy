## Реализация Задания 3.1b

```python
def paramFunc(t):
    xPoint = (3 * t**2 + 1) / (3 * t**3)
    yPoint = np.sin(t**3 / 3 + t)

    return [xPoint, yPoint]
```

```python
tMin, tMax = -10, 10;
tStep = (tMax - tMin) / 100
```

```python
#Создаем массив значений параметра t;
tArray = np.arange(tMin, tMax, tStep)

#Вычисляем координаты точек;
pointArray = paramFunc(tArray)

#Создаем матрицу, состоящую из пар (x, y);
matrix1b = np.transpose(pointArray)
```

```python
fig1b = plt.figure()
```

```python
#Максимальные и минимальные значения по осям Ox, Oy;
xMinB, xMaxB = np.min(pointArray[0]), np.max(pointArray[0])
yMinB, yMaxB = np.min(pointArray[1]), np.max(pointArray[1])

#Создам графическую область и задаем пределы по осям;
ax1b = plt.axes()
plt.axis([xMinB - 0.25, xMaxB + 0.25, yMinB - 10, yMaxB + 10])
```

```python
#Создание объекта;
line1b, = ax1b.plot([],[],'r:')
print(type(line1b))

line1b.get_xdata(), line1b.get_ydata()
```

```python
def atFrame1b(point):
 """
 Добавляет к объекту line1b точку с координатами (point[0],point[1])

 Arguments:
     1) point: массив из двух элементов
     2) Returns: None
 """

 xCoord = list(line1b.get_xdata())
 yCoord = list(line1b.get_ydata())

 xCoord.append(point[0])
 yCoord.append(point[1])

 line1b.set_data(xCoord, yCoord)
```

```python
#Создам графическую область;
fig1b = plt.figure()
ax1b = plt.axes()
plt.axis([xMinB - 0.25, xMaxB + 0.25, yMinB - 10, yMaxB + 10])

#Создаем графический объект типа Line2D;
line1b, = ax1b.plot([], [],'b', linewidth = 4)

#Создаем анимацию;
FuncAnimation(fig1b, atFrame1b, frames = matrix1b, repeat = False, interval = 0.1)
```

## Задание 3.2. Движение точки по линии на плоскости

```python
fig2 = plt.figure()
ax2 = plt.axes()
plt.axis([x_min-0.1, x_max+0.1, y_min-0.1, y_max+0.1])
'go' ,
line1, = ax2.plot(x_array, y_array,'b')  # объект для графика функции
line2, = ax2.plot([], [],'go') # объект для точки
```

```python
def at_frame2(point):
    """задает объект line2 точкой с координатами (point[0],point[1])
    Arguments :
    point : массив из двух элементов
    Returns : None
    """
    line2.set_data([point[0]],[point[1]])
```

```python
fig2 = plt.figure()
ax2 = plt.axes()
plt.axis([x_min-0.1, x_max+0.1, y_min-0.1, y_max+0.1])
def init():
    """создает начальное состояние графической области"""
    global line2 # переменная сделана глобальной, чтобы она могла изменяться вну
    line1, = ax2.plot(x_array, y_array,'b')  
    line2, = ax2.plot([], [],'go') 
    plt.legend([r'$y(x) = \sin(x)^5$','Мovable point'])
FuncAnimation(fig2, at_frame2, frames=matrix, init_func=init, 
        repeat=False, interval=20)
```

```python
#Создаем графическую область;
fig3 = plt.figure()
ax3 = plt.axes()
plt.axis([xMinB - 1, xMaxB + 1, yMinB - 1, yMaxB + 1])

def init():
    global pointB, secantLine, A

    curve, = ax3.plot(xArray, yArray, 'b')

    A = np.array([xArray[0], yArray[0]])
    pointA, = ax3.plot(A[0], A[1], 'ro')

    B = np.array([xArray[-1], yArray[-1]])
    pointB, = ax3.plot(B[0], B[1], 'go')

    secantP = [A + (B - A) * t for t in [-2, 2]]
    secantP = np.array(secantP)
    secantLine, = ax3.plot(secantP[:, 0], secantP[:, 1], 'g')

    plt.legend([r'$y(x) = \sin(x)^4$', 'Unmovable point A', 'Movable point B', 'Secant line'])

def atFrame3(point):
    """
    Задает объект pointB точкой с координатами (point[0],point[1]), а также меняет меняет положение
    secantLine, в зависимости от метоположения объекста pointB

     Arguments:
         1) point: массив из двух элементов, задающий координаты подвижной точки
         2) Returns: None
     """

    B = np.array([point[0], point[1]])
    pointB.set_data([point[0]], [point[1]])

    if np.allclose(A, B):
        secantLine.set_color('r')
        plt.legend([r'$y(x) = \sin(x)^4$', 'Unmovable point A', 'Movable point B', 'Tangent line'])
        
        x0, y0 = A[0], A[1]
        coeff = 4 * np.sin(x0) ** 3 * np.cos(x0)

        tangentP = np.array([A + np.array([1, coeff]) * t for t in [-100, 100]])
        secantLine.set_data(tangentP[:, 0], tangentP[:, 1])
    else:
        secantP = np.array([A + (B - A) * t for t in [-100, 100]])
        secantLine.set_data(secantP[:, 0], secantP[:, 1])

#Создание анимации движения точки по графику функции;
FuncAnimation(fig3, atFrame3, frames = matrix[::-1], init_func = init, repeat = False, interval = 10)
```
